```python
def debugger(data):
  """Allows for JS code to print to the ipynb"""
  print(data["data"])
  return {}

callbacks = {
    "debugger": debugger,
}
```

```python
import json

# The elements.json file defines the graph. This
# contains all nodes, information about each node,
# and edges.
with open("elements.json") as cy_elements_file:
  cy_elements_json = json.load(cy_elements_file)

# The style.json file contains the cytoscape style
# definition, which is similar to CSS. You can read
# more about it here: https://js.cytoscape.org/#style
with open("style.json") as cy_style_file:
  cy_style_json = json.load(cy_style_file)

# The layout.json file sets the positions of the
# nodes in the graph. Read more here:
# https://js.cytoscape.org/#layouts
with open("layout.json") as cy_layout_file:
  cy_layout_json = json.load(cy_layout_file)

data = {
    "elements": cy_elements_json,
    "style": cy_style_json,
    "layout": cy_layout_json,
}
```

```python
from notebookjs import execute_js

with open("../../bin/project-data-visualizations.js", "r") as f:
  cy_graph_js = f.read()

with open("./style.css", "r") as f:
  cy_graph_css = f.read()

css_list = [cy_graph_css, "https://unpkg.com/tippy.js@6/dist/tippy.css"]

execute_js(
		library_list=[cy_graph_js],
		main_function="projectdatavisualizations.drawGraph",
		data_dict=data,
		callbacks=callbacks,
		css_list=css_list,
)
```
